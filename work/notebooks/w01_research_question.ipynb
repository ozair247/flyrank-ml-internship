## 2. The Question: Decision, Action, and Cost of a Wrong Call

**The Decision:**  
Which specific pages should be prioritized for a content refresh to prevent traffic decline?

**The Action:**  
An SEO content specialist will take the top 50 ranked pages from the model's output and rewrite the headlines, update outdated statistics, and improve internal linking before moving to lower-ranked pages.

**The Cost of a Wrong Recommendation:**  
- **False Positive (Wasting time):** If the model recommends a page that is actually healthy, the team wastes ~2 hours editing content that didn't need it.  
- **False Negative (Missing a decline):** If the model misses a page that is about to crash, the client loses the ~731 median impressions (and potential clicks) that page was generating, leading to a direct loss in organic visibility and revenue.

**Why this isn't just "train a model":**  
This is a decision-support system, not a black-box prediction. The goal is to understand *why* a page is declining (low CTR, stale age, weak engagement) and provide a transparent, ranked list so a human makes the final call. The model serves the editor, it does not replace them.
